# Data Analysis 

## Packages 

```python
#Import necessary packages for data analysis/simulation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

##Import scipy to conduct normal distribution tests
from scipy import stats
```

## Data Cleaning 

```python
##Pull in data as a pandas dataframe
stocks = pd.read_csv('/content/drive/My Drive/sp500_joined_returns.csv')
##Pull in Binary y variable
y = pd.read_csv('/content/drive/My Drive/SPY.csv')

#set the index 
stocks.set_index(pd.to_datetime(stocks['Date']), inplace=True)
y.set_index(pd.to_datetime(y['Date']), inplace=True)

#drop unnecessary cols 
stocks.drop(['Unnamed: 0', 'Date'], axis=1, inplace=True)
y.drop(['Unnamed: 0', 'Date'], axis=1, inplace=True)
y.rename({'SPYUpDown': 'y'}, axis=1, inplace=True) 

##Merging our two set into 1 based on date 
merge = pd.merge(stocks,y, how='inner', left_index=True, right_index=True)
```

```python
merge.tail()
```

Becasue alot of the stocks have missing values we will only focus on ones who are not missing anything

```python
##Obtain list of company names present in the pandas dataframe
compNames = list(merge)

##Create an empty list for companies we will use
useComps = []

##Now determine which stocks do not have any missing values in the columns
for i in compNames:
    
    ##Boolean to asses if there are any missing values in the column 
    if merge[i].isnull().values.any() == False:
        
        ##If there are not missing values append to a new list
        useComps.append(i)
```

```python
len(useComps)-1
```

So we have 369 of the 500  orginal companies that have no missing data. Now lets remove all the other ones 

```python
stocks = merge[useComps]
```

```python
stocks.head()
```

## Data Analysis 

```python
##Create a subplot of the returns of all of our companies
##Resize the plotting canvas
plt.figure(figsize=(12,16))

##Loop to create out subplot
for i in range(60):
    
    ##Define the position we'd like to be in for the subplot
    plt.subplot(12,5,i+1)
    plt.title(useComps[i+1]+' Returns Histogram')
    stocks[useComps[i+1]].hist(bins=100)

##Adjust the layout of the subplot to avoid overlapping plots
plt.tight_layout()
```

So it looks like all of our priniciapl components appear to be noramal 

Now we need to select the stocks to use. S&P reports what the relative weight of each stock in the index is. We can use this to pick our optomial selection of stoicks. Clearly we want to pick the five that contribute the most the the price however we alos dont want there returns to be too correlated with eachother. We can use: 

MSFT: Microsoft Corporation - 5.65% (3rd) 
NVDA: NVIDIA Corporation - 1.43% (8th) 
UNH: UnitedHealth Group - 1.36% (9th) 
JNJ: Johnson and Johnson - 1.28% (11th) 
JPM: JPMorgan Chase and Co - 1.00% (13th) 


```python
comlist = ['MSFT', 'NVDA', 'UNH', 'JNJ', 'JPM']

sub_stocks = stocks[comlist]
```

```python
import seaborn as sns
corr = sub_stocks.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

We can see that we have sufficently low correlation amoung each stock so we feel confident that this subselection should be okay. 

```python
plt.rcParams['figure.figsize'] = [13, 10]
fig, axs = plt.subplots(3, 2, figsize=(12, 15))

axs[0,0].hist(sub_stocks['MSFT'], bins=100)
axs[0,0].set_title('MSFT')


axs[0,1].hist(sub_stocks['MSFT'], bins=100)
axs[0,1].set_title('MSFT')


axs[1,0].hist(sub_stocks['NVDA'], bins=100)
axs[1,0].set_title('NVDA')

axs[1,1].hist(sub_stocks['UNH'], bins=100)
axs[1,1].set_title('UNH')

axs[2,0].hist(sub_stocks['JNJ'], bins=100)
axs[2,0].set_title('JNJ')

axs[2,1].hist(sub_stocks['JPM'], bins=100)
axs[2,1].set_title('JPM')

fig.delaxes(axs[0,1]) 
```
